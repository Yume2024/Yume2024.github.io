## Tipos

Las variables están *debilmente tipadas* en python lo que significa que su tipo se deduce del contexto: la inicialización o los tipos de las variables utilizadas para su cálculo. Observa el siguiente ejemplo.

```python
print("Entero") # se imprime una cadena de caracteres
a = 3           # se define una variable con un valor de 3
print(a,type(a)) # se imprime la variable a con el tipo de dato de la variable

print("\nFloat") # se imprime un caracter nueva linea e imprime una cadena de caracteres
b = 3.14         # se define una variable con un valor de tipo float
print(b,type(b)) #se imprime la variable b con el tipo de dato de la variable

print("\nComplejo") # se imprime un caracter de nueva linea e imprime una cadena de caracteres
c = 3.14 + 2j       # se define una variable con una suma que contiene una parte real y imaginaria
print(c,type(c))    # se imprime la variable c con el tipo de variable que es en este caso complex ya que tiene una parte real y una imaginaria
print(c.real,type(c.real)) # se imprime la variable c pero solo su parte real
print(c.imag,type(c.imag)) # se imprime la variable c pero solo su parte imaginaria
```

Esta tipificación puede llevar a que alguna variable tenga tipos no deseados, lo que puede resolverse *casting*.

```python
d = 1j*1j        # se define una variable con una multiplicacion la cual es tipo complex la cal tiene una parte real y una imaginaria
print(d,type(d)) # se imprime la variable con el tipo de dato que es en este caso complex
d = d.real       # se difine la variable con su componente real del numero imaginario
print(d,type(d)) # se imprime la variable con su tipo de dato en este caso float por lo que solo es la parte real
d = int(d)       # se define la varible de tipo entero
print(d,type(d)) # se imprime la variable con el tipo de dato de la variable en este caso entero
```

```python
e = 10/3               # se define una variable donde diez se divide en tres
print(e,type(e))       # se imprime la variable con el tipo de varible que es la cual es tipo float ya que es el resultado de la division
f = (10/3)/(10/3)      # se define la variable f con dos divisones de la misma diez dividido tres
print(f,type(f))       # se imprime la variable f con el tipo de dato en este caso float
f = int((10/3)/(10/3)) # se define la varible f de tipo entero
print(f,type(f))       # se imprime la variable con el tipo de dato en este caso entero
```

## Listas

Las listas son el elemento base para secuencias de variables en python, son en sí mismas un tipo de variable.
* La sintaxis para escribirlas es `[ ... , ... ]`.
* Los tipos de los elementos pueden no ser todos iguales
* Los índices comienzan en $0$ (`l[0]` es el primer elemento de `l`)
* Las listas pueden estar anidadas (listas de listas de ...)


*Advertencia:* En Python existe otro tipo llamado *tupla* con la sintaxis `( ... , ... )`. Tiene casi la misma estructura que una lista con la notable excepción de que no se pueden añadir o eliminar elementos de una tupla. Las veremos brevemente más adelante.

```python
l = [1, 2, 3, [4,8] , True , 2.3] # Se define una lista
print(l, type(l)) # Se imprime la lista l y el tipo de dato de la lista.
```

```python
print(l[0],type(l[0])) # Imprime el primer elemento de la lista l y su tipo de dato.
print(l[3],type(l[3])) # Imprime el cuarto elemento de la lista l y su tipo de dato.
print(l[3][1],type(l[3][1])) # Imprime el segundo elemento de la lista que está dentro del cuarto elemento de la lista l, y su tipo de dato.
```

```python
print(l) # Imprime la lista completa l.
print(l[4:]) # Imprime los elementos de la lista l desde el índice 4 hasta el final.
print(l[:5]) # Imprime los elementos de la lista l desde el inicio hasta el índice 5 (sin incluirlo).
print(l[4:5]) # Imprime un segmento de la lista l desde el índice 4 hasta el índice 5 (sin incluirlo), lo cual en este caso solo imprime el elemento en el índice 4.
print(l[1:6:2]) # Imprime un segmento de la lista l desde el índice 1 hasta el índice 6 (sin incluirlo), con un paso de 2. Esto significa que imprime cada segundo elemento desde el índice 1 hasta el índice 6.
print(l[::-1]) # Imprime la lista l en orden inverso, comenzando desde el último elemento y avanzando hacia el primero.
print(l[-1]) # Imprime el último elemento de la lista l, utilizando un índice negativo que representa contar desde el final de la lista hacia atrás.
```

### Manejo de listas

Las listas son tipos tipo *puntero*. Esto significa que si escribes `l2=l`, no *copias* `l` a `l2` sino que copias el puntero, de modo que si modificas uno, modificarás el otro.

La forma correcta de copiar una lista es usar el método `copy` específico para variables de lista.

```python
l2 = l # Asigna la lista l a la variable l2.
l.append('Algo') # Agrega el valor Algo al final de la lista l utilizando el método append()
print(l,l2) # Imprime las listas l y l2. Dado que l2 hace referencia a la misma lista que l, ambos nombres apuntan a la misma lista modificada. Por lo tanto, imprimirán la misma lista con el valor 'Algo' agregado al final.
```

```python
l3 = list(l) # Crea una nueva lista l3 utilizando la función list() y pasando como argumento la lista l. Esto crea una copia superficial de la lista l, es decir, l3 ahora contiene los mismos elementos que l pero es una lista independiente.
l.remove('Algo') # Elimina la primera ocurrencia del valor 'Algo' de la lista l.
print(l,l3) # Imprime las listas l y l3. l ha sido modificada al eliminar 'Algo', pero l3 sigue siendo una copia de la lista original l
```

## Tuplas, Diccionarios [*]


* Las tuplas son similares a las listas pero se crean con `(...,...)` o simplemente comas. No se pueden modificar una vez creadas.

```python
t = (1,'b',876876.908) # Crea una tupla llamada t que contiene los elementos 1, 'b' y 876876.908.
print(t,type(t)) # Imprime la tupla t y su tipo de dato.
print(t[0]) # Imprime el primer elemento de la tupla t, que es 1.
```

```python
a,b = 12,[987,98987] # Asigna el valor 12 a la variable a y la lista [987, 98987] a la variable b. Aquí, a es una variable que contiene un entero, y b es una variable que contiene una lista.
u = a,b #  Crea una tupla llamada u que contiene las variables a y b. Esto crea una tupla con dos elementos, donde el primer elemento es el valor de a (que es 12) y el segundo elemento es el valor de b (que es la lista [987, 98987]).
print(a,b,u) # Imprime los valores de a, b y u.
```

```python
try: #  Este bloque intentará ejecutar el código dentro de él. En este caso, intentará modificar el valor de u[1].
    u[1] = 2 # Intenta asignar el valor 2 al segundo elemento de la tupla u. Sin embargo, esto generará un error porque las tuplas son inmutables y no se pueden modificar después de su creación.
except Exception as error: # Si ocurre alguna excepción dentro del bloque try, el flujo de control se desviará al bloque except. Exception as error captura la excepción que se produce y la asigna a la variable error.
    print(error) # Imprime el mensaje de error que se generó cuando se intentó modificar la tupla u.
```

* Los diccionarios están destinados a almacenar valores de la forma *clave-valor* con la sintaxis `{clave1 : valor1, ...}`.

Este tipo se utiliza a menudo como tipo de retorno en librerías.

```python
d = {"parametro1" : 1.0, "parametro2" : True, "parametro3" : "rojo"} # Crea un diccionario llamado d que contiene tres pares clave-valor
print(d,type(d)) # Imprime el diccionario d y su tipo de dato.
```

```python
print(d["parametro1"]) # Imprime el valor asociado con la clave "parametro1" en el diccionario d. En este caso, imprime 1.0, ya que ese es el valor inicial asociado con esa clave.
d["parametro1"] = 2.4 # Actualiza el valor asociado con la clave "parametro1" en el diccionario d. Se cambia el valor de 1.0 a 2.4.
print(d) # Imprime el diccionario d después de la modificación. Ahora, la clave "parametro1" tiene el valor actualizado de 2.4.
```

* Las cadenas se delimitan con comillas (dobles). Pueden manejarse globalmente del mismo modo que las listas (véase más arriba).

```python
print(s[0]) # Accede al primer carácter de la cadena s. En este caso, imprimirá "p", que es el primer carácter de la cadena "prueba".
print(s + "42") # Concatena la cadena s con la cadena "42" y luego imprime el resultado.
```

El método `formato`

```python
print( "test {}".format(42) ) # Esto crea una cadena de texto que dice "test ", seguida por {}, que es un marcador de posición donde se insertará el valor proporcionado por format(). Luego, se llama al método format(42) para insertar el valor 42 en el marcador de posición. Finalmente, esto devuelve la cadena completa "test 42" y Imprime la cadena de texto resultante, que es "test 42". El método format() reemplaza el marcador de posición {} con el valor proporcionado, que en este caso es 42.
```

```python
print( "Prueba con un int {:d}, un float {} (o {:e} que es aproximadamente {:.1f})".format(4 , 3.141 , 3.141 , 3.141 )) # Esto crea una cadena de texto que incluye marcadores de posición dentro de llaves {}. Estos marcadores de posición serán, El primer marcador de posición se reemplaza por el valor 4 como un entero (:d especifica que el valor debe ser formateado como un entero)., El segundo marcador de posición se reemplaza por el valor 3.141 como un flotante.,El tercer marcador de posición se reemplaza por el valor 3.141 como un flotante en notación científica (:e especifica que el valor debe ser formateado en notación científica).,El cuarto marcador de posición se reemplaza por el valor 3.141 como un flotante con un solo decimal ({:.1f} especifica que el valor debe ser formateado como un flotante con un solo decimal).
```

## If, Elif, Else

En Python, la formulación para la ramificación es la condición `if:` (atención al `:`) seguida de una sangría de *una tabulación* que representa lo que se ejecuta si la condición es verdadera. **La sangría es primordial y está en el núcleo de Python.

```python
enunciado1 = False # Se asigna el valor False.
enunciado2 = False # Se asigna el valor False.

if enunciado1: # Verifica si enunciado1 es True. Dado que enunciado1 es False, no se ejecuta el bloque de código dentro de este if.
    print("La sentencia1 es verdadera") # Imprime La sentencia1 es verdadera
elif enunciado2: #  Si enunciado1 es False, este bloque verifica si enunciado2 es True. Dado que enunciado2 también es False, no se ejecuta el bloque de código dentro de este elif.
    print("La sentencia2 es verdadera") # Imprime La sentencia1 es verdadera
else: # Si ninguno de los bloques if o elif anteriores se ejecuta, se ejecuta este bloque. En este caso, imprime "enunciado1 y enunciado2 son falsos", ya que ambos enunciados son False.
    print("enunciado1 y enunciado2 son falsos") # enunciado1 y enunciado2 son falsos
```

```python
enunciado1 = enunciado2 = True # Se asigna el valor True a ambas variables booleanas.

if enunciado1: # Verifica si enunciado1 es True. Dado que enunciado1 es True, el bloque de código dentro de este if se ejecuta.
    if enunciado2: # Verifica si enunciado2 es True. Dado que enunciado2 también es True, el bloque de código dentro de este segundo if se ejecuta.
        print("enunciado1 y enunciado2 son verdaderos") # Imprime "enunciado1 y enunciado2 son verdaderos", ya que tanto enunciado1 como enunciado2 son True.
```

```python
if enunciado1: # Comienza la estructura condicional externa, que verifica si enunciado1 es True. Dado que enunciado1 es True, el código dentro de este bloque if se ejecuta.
    if enunciado2: # Dentro del bloque if anterior, hay otra estructura condicional anidada que verifica si enunciado2 también es True. Dado que enunciado2 es True, el código dentro de este bloque if se ejecuta.
    #print("tanto enunciado1 como enunciado2 son Verdaderos") # Descomentar causaría un error. Esta línea está comentada y no afecta la ejecución del código. Si se descomentara, causaría un error de indentación porque no hay ninguna instrucción dentro del bloque if.
        print("aquí está bien") # Imprime "aquí está bien", ya que es una instrucción dentro del bloque if y se ejecuta correctamente.
    print("después de la línea anterior, aquí también") # Esta línea se ejecuta después del bloque if interno y está al mismo nivel de indentación que el bloque if interno, por lo que se ejecuta como parte del bloque if externo. Esto significa que se ejecuta después de la impresión "aquí está bien".
```

```python
enunciado1 = True # Se asigna el valor True a la variable enunciado1.

if enunciado1: #  verifica si enunciado1 es True. Dado que enunciado1 es True, el bloque de código dentro del if se ejecuta.
    print("se imprime si enunciado1 es True") # Imprime "se imprime si enunciado1 es True". Esta línea está dentro del bloque if y se ejecuta porque enunciado1 es True



    print("aún dentro del bloque if") # También imprime "aún dentro del bloque if". Aunque esta línea está más indentada que la primera, aún está dentro del bloque if y se ejecuta junto con la línea anterior. La indentación adicional se considera parte del bloque
```

```python
enunciado1 = False # Se asigna el valor False a la variable enunciado1.

if enunciado1: # verifica si enunciado1 es True. Dado que enunciado1 es False, el bloque de código dentro del if no se ejecuta.
    print("se imprime si enunciado1 es True") # Se imprime si la variable es True, pero como es False no entra en el bloque.

print("fuera del bloque if") # está fuera del bloque if, por lo que se ejecuta independientemente del valor de enunciado1. Esto significa que, aunque enunciado1 sea False, esta línea se imprimirá ya que está fuera del bloque condicional.
```

## Bucle For

La sintaxis de los bucles `for` es `for x in something:` seguido de una sangría de un tabulador que representa lo que se ejecutará.

El `algo` anterior puede ser de distinta naturaleza: lista, diccionario, etc.

```python
for x in [1, 2, 3]: # Este bucle for itera sobre cada elemento de la lista [1, 2, 3]. En cada iteración, el valor del elemento actual se asigna a la variable x.
    print(x) # En cada iteración del bucle, se imprime el valor de x
```

```python
sentence = "" # Se inicializa una cadena vacía llamada sentence, que se utilizará para construir la oración.
for word in ["Python", "for", "data", "Science"]: # Este bucle for itera sobre cada elemento de la lista
    sentence = sentence + word + " " # En cada iteración del bucle, la palabra actual (word) se concatena a la cadena sentence, seguida de un espacio en blanco. Esto construye gradualmente la oración.
print(sentence) # Una vez que se han concatenado todas las palabras en sentence, la oración resultante se imprime. Cada palabra estará separada por un espacio, formando una oración completa.
```

Una función útil es <tt>range</tt>, que genera secuencias de números que pueden utilizarse en bucles.

```python
print("Un tipo de rango") # Imprime el mensaje "Un tipo de rango".
for x in range(4): # Este bucle for itera sobre una secuencia de números enteros desde 0 hasta 3 (no incluido), generada por range(4).
    print(x) # imprime x, del 0-3

print("Otro tipo de rango") # Imprime el mensaje "Otro tipo de rango".
for x in range(2,6): # Este bucle for itera sobre una secuencia de números enteros desde 2 hasta 5 (no incluido), generada por range(2, 6).
    print(x) # imprime x, del 2-5

print("Y otro tipo de rango") # Imprime el mensaje "Y otro tipo de rango".
for x in range(1,12,3): # Este bucle for itera sobre una secuencia de números enteros desde 1 hasta 11 (no incluido), con un paso de 3, generada por range(1, 12, 3).
    print(x) # imprime el valor actual de x, que representa cada número en la secuencia generada por range(1, 12, 3). La secuencia es 1, 4, 7, y 10.
```

Si se necesita el índice junto con el valor, la función `enumerate` es útil.

```python
for idx, x in enumerate(range(-3,3)): # Este bucle for itera sobre una secuencia de números enteros desde -3 hasta 2 (no incluido), generada por range(-3, 3). La función enumerate() se utiliza para enumerar cada elemento de la secuencia junto con su índice. idx indice del for
    print(idx, x) # En cada iteración, imprime el índice (idx) y el valor (x) del elemento actual en la secuencia. La secuencia generada por range(-3, 3) es -3, -2, -1, 0, 1, y 2. Entonces, este bucle imprime cada índice y su correspondiente valor en la secuencia.
```

## Bucle while

De forma similar a los bucles `for`, la sintaxis es `while condition:` seguida de una sangría de un tabulador que representa lo que se ejecutará.

```python
i = 0 # Se inicializa la variable i con el valor 0.

while i<5: # Este bucle while se ejecutará mientras el valor de i sea menor que 5
    print(i) #  Imprime el valor actual de i.
    i+=1 # Incrementa el valor de i en 1 en cada iteración del bucle. Esto asegura que en cada iteración, i se incremente hasta que alcance o supere el valor de 5.
```

## Try [*]

Cuando un comando puede fallar, puedes `intentar` ejecutarlo y opcionalmente atrapar la `Exception` (es decir, el error).



```python
a = [1,2,3] # Se crea una lista a inicializada con los elementos [1, 2, 3].
print(a) # Imprime la lista a.

try: # En el primer bloque try, intenta modificar el valor en la posición 1 de la lista a a 3. Esto se hace mediante la asignación a[1] = 3. Dado que a es una lista mutable, esta operación es válida y no se produce ninguna excepción. Se imprime "Comando Ok".
    a[1] = 3 # asignación a[1] = 3
    print("Comando Ok") # Se imprime "Comando Ok".
except Exception as error: # se captura el error si existe
    print(error) # Se imprime el error

print(a) # Se imprime la lista a actualizada.

try: # intenta modificar el valor en la posición 6 de la lista a a 3. Sin embargo, la lista a solo tiene índices válidos hasta len(a) - 1, que es 2 en este caso. Intentar acceder o modificar un índice fuera de estos límites produce una excepción
    a[6] = 3 # modifica el valor en la posición 6 de la lista a a 3
    print("Comando Ok") # Se imprime "Comando Ok".
except Exception as error: # se captura el error si existe
    print(error) # Se imprime el error

print(a)
```

```python
def fun0(): # Esta línea define una función llamada fun0.
    print("\"fun0\" solo imprime") #  Esta línea imprime el mensaje "fun0" solo imprime.

fun0() # Esta línea llama a la función fun0().
```

Se puede añadir un docstring para documentar la función, que aparecerá al llamar a `help`.

```python
def fun1(l): # Esta línea define una función llamada fun1 que toma un parámetro l.
    """
    Imprime una lista y su longitud
    """
    print(l, " tiene una longitud ", len(l)) #  Imprime una lista pasada al parametro, y calcula su longitud

fun1([1,'iuoiu',True]) # se llama a la función fun1 con la lista [1, 'iuoiu', True] como argumento.
```

```python
help(fun1) # La función help() se utiliza para obtener ayuda sobre un objeto. Cuando se pasa una función como argumento a help(), devuelve información sobre esa función, incluyendo su documentación (docstring) si está disponible.
```

## Salidas

`return` da salida a una variable, tupla, diccionario, ...

```python
def square(x): # Esta línea define una función llamada square que toma un parámetro x.
    """
    Devuelve x al cuadrado.
    """
    return(x ** 2) #  Esta línea devuelve x elevado al cuadrado utilizando el operador **.

help(square) # imprimirá la documentación de la función square, que es la cadena de documentación (docstring) proporcionada en la definición de la función.
res = square(12) #  Llama a la función square con el argumento 12 y guarda el resultado en la variable res.
print(res) # Imprime el valor de res, que es el resultado de llamar a square(12)
```

```python
def powers(x): # Define una función llamada powers que toma un argumento x.
    """
    Devuelve las primeras potencias de x.
    """
    return(x ** 2, x ** 3, x ** 4) # Retorna una tupla que contiene las primeras tres potencias de x (al cuadrado, al cubo y a la cuarta potencia).

help(powers) # Llama a la función help() para obtener información sobre la función powers
```

```python
res = powers(12) # Llama a la función powers con el argumento 12, lo que devuelve una tupla que contiene las primeras tres potencias de 12 (al cuadrado, al cubo y a la cuarta potencia). Este resultado se asigna a la variable res.
print(res, type(res)) # Imprime el valor de res, que es la tupla devuelta por la función powers cuando se llamó con el argumento 12. Luego, imprime el tipo de res usando la función type(). Esto muestra que res es de tipo tuple.
```

```python
two,three,four = powers(3) # Llama a la función powers con el argumento 3, lo que devuelve una tupla que contiene las primeras tres potencias de 3 (al cuadrado, al cubo y a la cuarta potencia). Luego, desempaqueta esta tupla en las variables two, three y four. two contendrá la primera potencia (3 al cuadrado), three contendrá la segunda potencia (3 al cubo) y four contendrá la tercera potencia (3 a la cuarta potencia).
print(three,type(three)) # Imprime el valor de three, que es la segunda potencia de 3 (3 al cubo). Luego, imprime el tipo de three utilizando la función type(). Esto muestra que three es de tipo int (entero).
```

```python
def powers_dict(x): # Define una función llamada powers_dict que toma un argumento x.
    """
    Devuelve las primeras potencias de x como un diccionario.
    """
    return{"two": x ** 2, "three": x ** 3,  "four": x ** 4} # Devuelve un diccionario que contiene las primeras tres potencias de x (al cuadrado, al cubo y a la cuarta potencia), utilizando las claves "two", "three" y "four". La cadena de documentación (docstring) proporciona información sobre lo que hace la función.


res = powers_dict(12) # Llama a la función powers_dict con el argumento 12, lo que devuelve un diccionario que contiene las primeras tres potencias de 12 (al cuadrado, al cubo y a la cuarta potencia). Este resultado se asigna a la variable res.
print(res, type(res)) # Imprime el valor de res, que es el diccionario devuelto por la función powers_dict cuando se llamó con el argumento 12. Luego, imprime el tipo de res utilizando la función type(). Esto muestra que res es de tipo dict (diccionario).
print(res["two"],type(res["two"])) # Imprime el valor asociado a la clave "two" en el diccionario res, que es la segunda potencia de 12 (12 al cuadrado). Luego, imprime el tipo de este valor utilizando la función type(). Esto muestra que el valor es de tipo int (entero).
```

## Argumentos

Es posible
* Dar los argumentos en cualquier orden siempre que se escriba el nombre de la variable argumental correspondiente
* Establecer valores por defecto a las variables para que se conviertan en opcionales

```python
def fancy_power(x, p=2, debug=False): # Define una función llamada fancy_power que toma tres parámetros.
    """
    He aquí una versión elegante de potencia que calcula el cuadrado del argumento u otras potencias si se establece p
    """
    if debug: # Si el parámetro debug está configurado como True, imprime un mensaje que indica que la función fancy_power se está llamando con los valores de x y p.
        print( "\"fancy_power\" se llama con x =", x, " y p =", p)
    return(x**p) # Retorna el resultado de elevar x a la potencia p.
```

```python
print(fancy_power(5)) # Llama a la función fancy_power con el argumento 5. Dado que no se proporciona un valor para el parámetro p, la función utilizará el valor predeterminado de p, que es 2. Por lo tanto, calculará el cuadrado de 5. No se especifica el parámetro debug, por lo que no se imprimirá ninguna información de depuración.
print(fancy_power(5,p=3)) # Llama a la función fancy_power con el argumento 5 y el parámetro p establecido en 3. Esto indica que la función debe calcular 5 elevado a la tercera potencia. No se especifica el parámetro debug, por lo que no se imprimirá ninguna información de depuración.
```

```python
res = fancy_power(p=8,x=2,debug=True)# Llama a la función fancy_power con los siguientes argumentos y parámetros opcionales: p=8,x=2,debug=True
print(res) # Imprime el valor de res, que es el resultado devuelto por la función fancy_power
```

# 4- Clases [*]


Las clases son el núcleo de la programación *orientada a objetos*, se utilizan para representar un objeto con *atributos* (variables) y *métodos* (funciones) relacionados.

Se definen como funciones pero con la palabra clave class `clase mi_clase(objeto):` seguida de una sangría. La definición de una clase suele contener algunos métodos:
* El primer argumento de un método debe ser `self` en auto-referencia.
* Algunos nombres de métodos tienen un significado específico:
   * `__init__`: método ejecutado en la creación del objeto.
   * `__str__`: método ejecutado para representar el objeto como una cadena, por ejemplo cuando el objeto se pasa a la función `print`.

```python
class Point(object): # Este código define la clase Point
    """
    Clase de un punto en el plano 2D.
    """
    def __init__(self, x=0.0, y=0.0): #  El constructor de la clase, que inicializa las coordenadas x e y del punto. Por defecto, el punto se crea en la posición (0.0, 0.0), pero también se pueden proporcionar valores específicos para x e y.
        """
        Creación de un nuevo punto en la posición (x, y).
        """
        self.x = x # asigna el valor de self.x con el valor por parametro
        self.y = y # asigna el valor de self.y con el valor por parametro

    def translate(self, dx, dy): # Un método para trasladar el punto en las direcciones x e y proporcionadas.
        """
        Traslada el punto por (dx , dy).
        """
        self.x += dx # toma el valor actual de self.x y le suma por parametro dx
        self.y += dy # toma el valor actual de self.y y le suma por parametro dy

    def __str__(self): # Se define el método __str__
        return("Punto: ({:.2f}, {:.2f})".format(self.x, self.y)) # devuelve una cadena formateada que representa las coordenadas del punto con dos decimales de precisión.
```

```python
p1 = Point() # Crea un nuevo objeto Point sin especificar coordenadas, por lo que se inicializa en la posición predeterminada (0.0, 0.0).
print(p1) # Imprime la representación en forma de cadena de p1

p1.translate(3,2) # Llama al método translate en el objeto p1, trasladando sus coordenadas por 3 unidades en la dirección x y 2 unidades en la dirección y.
print(p1) # Imprime la representación en forma de cadena de p1

p2 = Point(1.2,3) # Crea un nuevo objeto Point con coordenadas especificadas (1.2, 3).
print(p2) # Imprime la representación en forma de cadena de p2
```
